```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
full = cv2.imread('../image/d_full.jpg', cv2.IMREAD_COLOR)
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
plt.imshow(full)
```

```python
face = cv2.imread('../image/d_face.jpg', cv2.IMREAD_COLOR)
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
plt.imshow(face)
```

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',' cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    # CREATE COPY OF FULL IMAGE
    full_copy = full.copy()
    
    method = eval(m)
    
    # TEMPLATE MATCHING
    res = cv2.matchTemplate(full_copy, face, method)
    
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
        
    height, width, channel = face.shape
    
    bot_right = (top_left[0]+width, top_left[1]+height)
    
    cv2.rectangle(full_copy, top_left, bot_right, (255,0,0), 15)
    
    # PLOT IMAGES 
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title('Heat map')
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title('Detection of template')
    
    plt.suptitle(m)
    
    plt.show()
    
    print('\n')
```
