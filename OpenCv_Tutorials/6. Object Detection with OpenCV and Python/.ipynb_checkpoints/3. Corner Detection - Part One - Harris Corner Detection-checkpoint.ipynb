```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../image/chess.jpg')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.imread('../image/chess.jpg', 0)
plt.imshow(gray_flat_chess, cmap='gray')
```

```python
real_chess = cv2.imread('../image/real_chess.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
plt.imshow(real_chess)
```

```python
gray_real_chess = cv2.imread('../image/real_chess.jpg', 0)
plt.imshow(gray_real_chess, cmap='gray')
```

```python
gray = np.float32(gray_flat_chess)
```

```python
gray
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=11, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
plt.imshow(dst, cmap='gray')
```

```python
flat_chess[dst > 0.01 * dst.max()] = [255, 0, 0] # RGB
```

```python
plt.imshow(flat_chess)
```

```python
gray = np.float32(gray_real_chess)
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
plt.imshow(dst)
```

```python
real_chess[dst > 0.01*dst.max()] = [255, 0, 0] # rgb
```

```python
plt.imshow(real_chess)
```
