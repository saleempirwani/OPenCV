```python
import cv2
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def disp(img, cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
resses = cv2.imread('../image/resses.jpg', 0)
disp(resses)
```

```python
cereals = cv2.imread('../image/cereals.jpg', 0)
disp(cereals)
```

```python
sift = cv2.__version__
sift

#!pip uninstall opencv-python
#!pip install opencv-contrib-python==3.4.2.16
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(resses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher()
```

```python
matches = bf.knnMatch(des1, des2, k=2)
```

```python
len(matches)
```

```python
good = []

# lESS DISTANCE == BETTER MATCH
# RATIO MATCH1 < 75% MATCH 2

for m1, m2 in matches:
    # IF MATCH 1 DISTANCE IS LESS THAN 75% OF MATCH 2 DISTANCE
    # THEN DESCRIPTOR WAS A GOOD MATCH, LETS KEEP IT
     if m1.distance < 0.75 * m2.distance:
            good.append([m1])
```

```python
len(good)
```

```python
sift_matches = cv2.drawMatchesKnn(resses, kp1, cereals, kp2, good, None, flags=2)
```

```python
disp(sift_matches)
```

## Flaynn Matches

```python
resses = cv2.imread('../image/resses.jpg', 0)
disp(resses)
```

```python
cereals = cv2.imread('../image/cereals.jpg', 0)
disp(cereals)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1, des1 = sift.detectAndCompute(resses, None)
kp2, des2 = sift.detectAndCompute(cereals, None)
```

```python
# FLANN
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
search_params = dict(checks=50)
```

```python
flann = cv2.FlannBasedMatcher(index_params, search_params)
```

```python
matches = flann.knnMatch(des1, des2, k=2)
```

```python
matchesMask = [[0,0] for i in range(len(matches))]
# matchesMask
```

```python
# good = []

# for m1, m2 in matches:
#     if m1.distance < 1 * m2.distance:
#         good.append([m1])
for i, (m1, m2) in enumerate(matches):
    if m1.distance < 0.70 * m2.distance:
        matchesMask[i] = [1, 0]
```

```python
draw_params = dict(matchColor=(0,255,0),
                  singlePointColor=(255,0,0),
                  matchesMask=matchesMask,
                  flags=0)
```

```python
# flann_matches = cv2.drawMatchesKnn(resses, kp1, cereals, kp2, good, None, flags=0)
flann_matches = cv2.drawMatchesKnn(resses, kp1, cereals, kp2, 
                                   matches, None, **draw_params)

disp(flann_matches)
```
