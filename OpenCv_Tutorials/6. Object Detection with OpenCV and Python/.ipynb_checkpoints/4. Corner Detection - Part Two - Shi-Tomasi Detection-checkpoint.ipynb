```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread('../image/chess.jpg')
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)

real_chess = cv2.imread('../image/real_chess.jpg')
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)

gray_flat_chess = cv2.imread('../image/chess.jpg', 0)
gray_real_chess = cv2.imread('../image/real_chess.jpg', 0)
```

```python
corners = cv2.goodFeaturesToTrack(gray_flat_chess, 64, 0.01, 10)
# corners
```

```python
corners = np.int0(corners)
```

```python
corners
```

```python
for i in corners:
    x,y = i.ravel() # to convert in One - dim
    cv2.circle(flat_chess, (x,y), 5, (255, 0, 0), -1)
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(gray_real_chess, 120, 0.01, 10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess, (x,y), 3, (255, 0, 0), -1)
```

```python
plt.imshow(real_chess)
```
