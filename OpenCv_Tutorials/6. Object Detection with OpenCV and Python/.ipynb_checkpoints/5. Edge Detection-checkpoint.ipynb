```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../image/d_face.jpg')
plt.imshow(img)
```

```python
edge = cv2.Canny(img, threshold1=127, threshold2=127)
plt.imshow(edge)
```

```python
edge = cv2.Canny(img, threshold1=0, threshold2=255)
plt.imshow(edge)
```

```python
med_val = np.median(img)
med_val
```

```python
# LOWER THRESHOLD EITHER 0% OR 70% OF THE MEDIAN VALUE, WHICHEVER IS GREATER
lower = int(max(0, 0.7 * med_val))
print('lower', lower)
# UPPER THRESHOLD EITHER 130% OF MEDIAN VALUE OR THE MAX 255, WHICEVER IS LOWER
upper = int(min(255, 1.3 * med_val))
print('upper', upper)
```

```python
edge = cv2.Canny(img, threshold1=lower, threshold2=upper)
plt.imshow(edge)
```

```python
blur_img = cv2.blur(img, ksize=(4,4))
edge = cv2.Canny(blur_img, threshold1=lower, threshold2=upper+20)
plt.imshow(edge)
```
