```python
import cv2
import numpy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def disp(img, cmap='gray'):
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
resses = cv2.imread('../image/resses.jpg', 0)
disp(resses)
```

```python
cereals = cv2.imread('../image/cereals.jpg', 0)
disp(cereals)
```

```python
orb = cv2.ORB_create()
```

```python
kp1, des1 = orb.detectAndCompute(resses, None)
kp2, des2 = orb.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
matches = bf.match(des1, des2)
```

```python
matches = sorted(matches, key=lambda x:x.distance )
```

```python
m = matches[158]
m.distance
```

```python
resses_matches = cv2.drawMatches(resses, kp1, cereals, kp2, matches[:25], None, flags=2)
```

```python
disp(resses_matches)
```
