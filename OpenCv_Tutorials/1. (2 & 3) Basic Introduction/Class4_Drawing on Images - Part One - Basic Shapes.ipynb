```python
cv2.line(img=blank_img, pt1=(0,0), pt2=(512,512), color=(255, 255, 255), 
        thickness=5, lineType=cv2.LINE_8)
plt.imshow(blank_img)
```

### Class5_Drawing on Images Part Two - Text and Polygons

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Biliyorum', org=(10, 500), fontFace=font, 
            fontScale=2, color=(255, 255, 0), thickness=3, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

#### Polygon (your desired shape)

```python
ver = np.array([ [100,300], [200,200], [400,300], [200,400] ], np.int32)
```

```python
ver
```

```python
ver.shape
```

```python
pts = ver.reshape((-1, 1, 2)) 
```

```python
pts.shape
```

```python
pts.ndim
```

```python
pts
```

```python
blank_img = np.zeros((512, 512))
cv2.polylines(blank_img, [pts], isClosed=True, color=(255, 0, 255), thickness=5)
plt.imshow(blank_img)
```

```python
ver = np.array([ [100,400], [400,400], [250,100]], dtype=np.int32)
pt = ver.reshape((3, 1, -1))
print(pt.shape)
blank_img = np.zeros((512, 512))
cv2.polylines(blank_img, [pt], isClosed=True, color=(255, 255, 255),
             thickness=5)
plt.imshow(blank_img)
```

```python
ver = np.array([ [100,400], [400,400], [250,100]], dtype=np.int32)
pt = ver.reshape((3, 1, -1))
print(pt.shape)
blank_img = np.zeros((512, 512))
cv2.fillPoly(blank_img, [pt], color=(255, 255, 255))
plt.imshow(blank_img)
```
