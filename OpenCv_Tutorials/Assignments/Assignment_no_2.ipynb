### Draw empty red rectangle around dog face and display it in notebook

```python
img = cv2.imread('image/puppy1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.rectangle(img, pt1=(95, 10), pt2=(800, 580), color=(255, 0, 0), thickness=5)
plt.imshow(img)
```

### Draw empty blue triangle around dog face and display it in notebook

```python
img = cv2.imread('image/puppy1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#ver = np.array([[440,10], [10,580], [670,580]])
ver = np.array([[370,10], [40,580], [700,580]])
ver = ver.reshape((-1, 1, 2))
cv2.polylines(img, [ver], isClosed=True, color=(0, 255, 255), thickness=5)
plt.imshow(img)
```

### Draw filled blue triangle around dog face and display it in notebook

```python
img = cv2.imread('image/puppy1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ver = np.array([[370,10], [40,580], [700,580]])
ver = ver.reshape((-1, 1, 2))
cv2.fillPoly(img, [ver], color=(255, 255, 0))
plt.imshow(img)
```

```python
def draw_circle(event, x, y, flags, params):
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, center=(x,y), radius=50, color=(255, 255, 255), 
                   thickness=5)
        
cv2.namedWindow(winname='circle')
cv2.setMouseCallback('circle', draw_circle)

img = cv2.imread('image/puppy1.jpg')
while True:
    cv2.imshow('circle', img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
        
cv2.destroyAllWindows()
```
