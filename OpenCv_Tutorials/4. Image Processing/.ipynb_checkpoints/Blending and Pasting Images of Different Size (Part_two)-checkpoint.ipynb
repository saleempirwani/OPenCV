### Blending and Pasting Images of Different Size (Part_two)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('../image/puppy1.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('../image/gg.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2 = cv2.resize(img2, (250,250))
img2.shape
```

```python
plt.imshow(img2)
```

```python
plt.imshow(img1)
```

```python
img1.shape
```

```python
x_off = 1023 - 250
y_off = 812 - 250
```

```python
img2.shape
```

```python
row, col, chn = img2.shape
```

```python
roi = img1[y_off:812, x_off:1023]
plt.imshow(roi)
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
plt.imshow(img2gray, cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
plt.imshow(mask_inv, cmap='gray')
```

```python
mask_inv.shape
```

```python
white_back = np.full(img2.shape, 255, dtype=np.uint8)
```

```python
white_back
white_back.shape
```

```python
bk = cv2.bitwise_or(white_back, white_back, mask=mask_inv)
```

```python
bk.shape
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi, fg)
plt.imshow(final_roi)
```

```python
large_img = img1
small_img = final_roi
```

```python
large_img[y_off:y_off+small_img.shape[0], x_off:x_off+small_img.shape[1]] = small_img
```

```python
plt.imshow(large_img)
```
