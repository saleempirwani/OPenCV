```python
img = brick
```

```python
color = ('b', 'g', 'r')
```

```python
for i,col in enumerate(color):
    hist = cv2.calcHist([img], channels=[i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])
plt.title('histrogram for blue bricks')
```

```python
for i,col in enumerate(color):
    hist = cv2.calcHist([horse], channels=[i], mask=None, histSize=[256], ranges=[0,256])
    plt.plot(hist, color=col)
    plt.xlim([0,50])
    plt.ylim([0, 5000])
plt.title('histrogram for horse')
```
