### Blending and Pasting Images of Same Size (Part_one)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('../image/puppy1.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.imread('../image/gg.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
img1 = cv2.resize(img1, (600,600))
img2 = cv2.resize(img2, (600,600))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blending = cv2.addWeighted(src1=img1, alpha=0.5, src2=img2, beta=0.5, gamma=0)
plt.imshow(blending)
```
