### Thresholding
1. Seperation of light and dark areas
2. Suppose threshold value = 60, if grey scale image values exceeds this threshold then it may consider to light area(255 or binary 1) otherwise dark region (0, or binary 0).

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../image/rainbow.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
img.max()
```

```python
# img if img > thresh; 0 otherwise
ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
```

```python
ret
```

```python
plt.imshow(thresh, cmap='gray')
```

```python
img = cv2.imread('../image/crossword.jpg', 0)
```

```python
plt.imshow(img, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
show_pic(img)
```

```python
# maxvalue if img > thresh; 0 otherwise
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 8)
show_pic(th2)
```

### Blended
1. overlay 2 image with a specified opacity

```python
blended = cv2.addWeighted(src1=th1, alpha=0.5, src2=img, beta=0.5, gamma=0)
show_pic(blended)
```
