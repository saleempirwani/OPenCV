```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
gorilla = cv2.imread('../image/gorilla.jpg', 0)
```

```python
display(gorilla, cmap='gray')
```

```python
gorilla.shape
```

```python
hist = cv2.calcHist([gorilla], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(hist)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
display(eq_gorilla, cmap='gray')
```

```python
hist = cv2.calcHist([eq_gorilla], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
plt.plot(hist)
```
