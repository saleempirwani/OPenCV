```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rainbow = cv2.imread('../image/rainbow.jpg') # BGR
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB) # RGB
```

```python
img = rainbow
```

```python
img.shape
```

```python
img.shape[:2]
```

```python
mask = np.zeros(img.shape[:2], np.uint8)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
mask[600:800, 400:1200] = 255
```

```python
plt.imshow(mask, cmap='gray')
```

```python
plt.imshow(img)
```

```python
masked_img = cv2.bitwise_and(img, img, mask=mask)
```

```python
show_masked_img = cv2.bitwise_and(show_rainbow, show_rainbow, mask=mask)
```

```python
plt.imshow(show_masked_img)
```

```python
hist_mask_val_red = cv2.calcHist([rainbow] , channels=[2], mask=mask, histSize=[256], ranges=[0,256])
```

```python
hist_val_red = cv2.calcHist([rainbow] , channels=[2], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_mask_val_red)
plt.title('hist_mask_val_red')
```

```python
plt.plot(hist_val_red)
plt.title('hist_val_red')
```
